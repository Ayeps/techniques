## Background

When customer use the APP to apply a personal loan, the customer should access his/her mobile contacts list to the APP. Therefore, the topic is whether we could utilize the contacts information to support creadit risk recognition. Concretely, could we recognize a fraud loan application if the mobile phone is purposed equipment for fraudulency, or could we recogize the application is from a loan intermediary agency? So, that is the purpose we do this exploration.

The analysis includes the basic machine learning knowledge such as **Nature Language Process** and **Anomaly Detection**, and it is done by **Python**. However, if you are not famillar with the algorithm or python itself, it doesn't matter. We have step by step description to illustrate our thoughts.

## Text segmentation and part of speech Tagging

Since the contacts list is generated by customer, every customer has its own pattern to store the contact name. For example, you could directly use the name of your relatives as the contacts name or you could just use the appellation. Moreover, it is easier to observe that there are lots of words like "xx客户" and company names in the contacts when the list is from sales staff. But for common cases, we assume that the real person names dominate the list. So the idea is whether we could identify those non-common list from the majority. Concretely, we want to identify the meaning words as well as people names in the contact list and then we classify the common and non-common lists.

So the problem is how we could extract the features from the contacts or the chinese names. For example, if one has a contact name of "北京大学张奕老师", first of all, we have to understand that "北京大学张奕老师" is only one word or several words together. It is different from English since in English sentence each word is divided by a space. But it is not the case of Chinese. When comes to Chinese, text segmentation should be the first step. For "北京大学张奕老师", it is perfect if we could use some text segmentation algorithm to get the seperate words like "北京大学" , "张奕" and "老师". However, it is still good if we got "北京", "大学" , "张奕" and "老师". Here, we will not discuss more about text segmentation, since it is another intereting topic. And also in **Python**, **R**, **Java**, there are lots of useful library which implement the algorithm you just not worry about that. 

After text segmentation, for each word, we tag the part of speech using labels compatible with ictclas(Institute of Computing Technology, Chinese Lexical Analysis System). Part of speech Tagging is easier for chinese, since majority of words have only one part of speech or frequency of the second part of speech is quite lower than the first.

In this analysis, we use [BosonNLP(Python)](http://bosonnlp.com/) to implement Text segmentation and POS tagging. For POS tagging rule, refer to [POS tagging](http://docs.bosonnlp.com/tag_rule.html).

```python
# -*- coding: utf-8 -*-
from bosonnlp import BosonNLP
from __future__ import unicode_literals
from collections import Counter
import os
nlp = BosonNLP('pPycccDe.4574.jJ3uiPlwQ9ln')

dc2['name_cut'] = dc2['mobile_name'].apply(lambda x:nlp.tag(x))
dc2['tag_cnt'] = dc2['name_cut'].apply(lambda x:Counter(x[0]['tag']))

dc2['tag_n'] = dc2['tag_cnt'].apply(lambda x: x['n'])
dc2['tag_nr'] = dc2['tag_cnt'].apply(lambda x: x['nr'])
dc2['tag_ns'] = dc2['tag_cnt'].apply(lambda x: x['ns'])
dc2['tag_nt'] = dc2['tag_cnt'].apply(lambda x: x['nt'])
dc2['tag_nz'] = dc2['tag_cnt'].apply(lambda x: x['nz'])
dc2['tag_nl'] = dc2['tag_cnt'].apply(lambda x: x['nl'])
dc2['tag_t'] = dc2['tag_cnt'].apply(lambda x: x['t'])
dc2['tag_s'] = dc2['tag_cnt'].apply(lambda x: x['s'])
dc2['tag_f'] = dc2['tag_cnt'].apply(lambda x: x['f'])
dc2['tag_v'] = dc2['tag_cnt'].apply(lambda x: x['v'])
dc2['tag_a'] = dc2['tag_cnt'].apply(lambda x: x['a'])
dc2['tag_b'] = dc2['tag_cnt'].apply(lambda x: x['b'])
dc2['tag_z'] = dc2['tag_cnt'].apply(lambda x: x['z'])
dc2['tag_r'] = dc2['tag_cnt'].apply(lambda x: x['r'])
dc2['tag_m'] = dc2['tag_cnt'].apply(lambda x: x['m'])
dc2['tag_q'] = dc2['tag_cnt'].apply(lambda x: x['q'])
dc2['tag_d'] = dc2['tag_cnt'].apply(lambda x: x['d'])
dc2['tag_p'] = dc2['tag_cnt'].apply(lambda x: x['p'])
dc2['tag_c'] = dc2['tag_cnt'].apply(lambda x: x['c'])
dc2['tag_u'] = dc2['tag_cnt'].apply(lambda x: x['u'])
dc2['tag_y'] = dc2['tag_cnt'].apply(lambda x: x['y'])
dc2['tag_o'] = dc2['tag_cnt'].apply(lambda x: x['o'])
dc2['tag_nx'] = dc2['tag_cnt'].apply(lambda x: x['nx'])
dc2['tag_w'] = dc2['tag_cnt'].apply(lambda x: x['w'])

dc2.to_csv('F:\\data\\contacts\\contacts.csv',encoding='utf-8')
```

With the algorithm of Text segmentation and part of Speech Tagging, it is possible to recognize the pattern of the contacts. For example, if user have one contact named "万昕", we will easily reconized "万昕" as "nr"(nr stands for 人名). If named "万昕老师", the alogrithm will cut it into 2 words with tagging 万昕/nr（人名） and 老师/n（名词）.

We have to say that all the text segmentation and tagging algorithm is not perfect. For if we give the word "趣分期", it will be cut into "趣" and "分期" as a noun and a adverb. It is because the dictionary we used to text segmentation has not a word like "趣分期". But we could design the adapitive dictionary to learn the new words with some feedback loop.  

```python
result = nlp.tag('万昕')
for d in result:
    print(' '.join(['%s/%s' % it for it in zip(d['word'], d['tag'])]))

result = nlp.tag('万昕老师')
for d in result:
    print(' '.join(['%s/%s' % it for it in zip(d['word'], d['tag'])]))
    
result = nlp.tag('客户_张秋奕')
for d in result:
    print(' '.join(['%s/%s' % it for it in zip(d['word'], d['tag'])]))
    
result = nlp.tag('趣分期')
for d in result:
    print(' '.join(['%s/%s' % it for it in zip(d['word'], d['tag'])]))
```

## Application of contacts tagging feature

### Case 1: Anomaly detection by Gaussian Mixture Models

When we scan the customers' contact list, we realize that it exists some abnormal contact list as below:
* The proportion of real person name is small in the list. e.g. "分期乐, 圆通快递, 机子, 仁,可可,技术打,吴先生,吴先生,张明,北京贷款，货款，北京"
* Person name with too much noun in the list, which indicates the mobile is used for some business. e.g. "王云客户,刘胜客户,陈庆齐客户,许先生,李旭客户,红红,王张红,王晓燕,丁江客户"

So we check the distribution of nr(name) proportion and n(noun) proportion as below. And it shows that either nr and n proportion seems to obey the Gaussian Distribution. According to **Central limit theorem**, we make the Gaussian Distribution assumption for n and nr. From the charts, however, in this sample, it indicates there are some outliers or anomaly points in the data to shift the mean value. So far we could generate some assumptions as below:

+ The majority of the customer contacts lists are normal
+ The normal point's nr and n proportion obey Gaussian Distribution

```python
ax_list = df_init_tag.hist(column=['tag_n','tag_nr'],figsize=(15, 30),bins=np.linspace(0,1,50),layout=(10,3))
for ax_item in ax_list.flat:
    ax_item.set_xlim(0,1)
    try:
        #ax_item.set_title(unicode(plot_title_dict[ax_item.get_title()],'utf-8'), fontproperties=font)
        ax_item.set_title(plot_title_dict[ax_item.get_title()], fontproperties=font)
    except:
        continue

print 'E(n) = %f, Var(n) = %f' %(df_init_tag['tag_n'].mean(),df_init_tag['tag_n'].var())
print 'E(nr) = %f, Var(nr) = %f' %(df_init_tag['tag_nr'].mean(),df_init_tag['tag_nr'].var())
```

So the problem is defined as Anomaly Detection for such abnormal customer contacts list. We hope we could detect these customer when in the underwriting stage.

As previous assumption that it supposes that nr and n proportion obey (Multi)Gaussian Distribution, so Density Estimation is used to estimate Gaussian Parameter(expectation and variance), as well as probability density function. For simplicity, here we use Gaussian Mixture Models with one component. Then the hard or soft threshold could be used to decide which point is abnormal. Concretely, the point far away from central of probablity density should be anomaly or outlier compared with majority.

As below, **Fig1** is the scatter of tag_n and tag_nr. **Fig 2 and Fig 3** shows probability density surface in 3-D dimension. 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from sklearn import mixture
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter


fig = plt.figure(figsize=(15,20))

# fit a Gaussian Mixture Model with one components
X_train = df_init_tag[['tag_n','tag_nr']].values
clf = mixture.GMM(n_components=1, covariance_type='full')
clf.fit(X_train)

# display predicted scores by the model as a surface plot
x = np.linspace(0, 1.0)
y = np.linspace(0, 1.0)
X, Y = np.meshgrid(x, y)
XX = np.array([X.ravel(), Y.ravel()]).T
ZZ = np.exp(clf.score_samples(X_train)[0])
Z = np.exp(clf.score_samples(XX)[0])
Z = Z.reshape(X.shape)

ax = fig.add_subplot(3,1,1)
plt.title('Fig1: Tag_n/Tag_nr Scatter')
ax.scatter(x=X_train[:,0],y=X_train[:,1])
ax.set_xlabel('%名词', fontproperties=font)
ax.set_ylabel('%人名', fontproperties=font)
ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.invert_xaxis()
ax.invert_yaxis()

ax = fig.add_subplot(3,1,2,projection='3d')
plt.title('Fig2: Log-likelihood predicted by a Gaussian Mixture Models')
ax.set_xlabel('%名词', fontproperties=font)
ax.set_ylabel('%人名', fontproperties=font)
ax.scatter3D(X_train[:,0],X_train[:,1],ZZ)
ax.set_ylabel('%人名', fontproperties=font)
ax.view_init(45,45)
ax.set_xlim(0,1,emit=True)
ax.set_ylim(0,1,emit=True)
ax.set_zlim(0,5)


ax = fig.add_subplot(3,1,3,projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
#fig.colorbar(surf, shrink=0.5, aspect=5)
plt.title('Fig3: Probabiliy Density Surface of GMMs')
ax.set_xlabel('%名词', fontproperties=font)
ax.set_ylabel('%人名', fontproperties=font)
ax.view_init(45,45)
ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_zlim(0,5)


plt.show()
```

For instance, if we use a plane(prob_cut=0.5) to cut the probability density, we could classify the anomaly(BLUE) and normal point(RED).

```python
prob_cut = 0.5

fig = plt.figure(figsize=(15,20))

ax = fig.add_subplot(2,1,1,projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
surf1 = ax.plot_surface(X, Y, np.ones(X.shape)*prob_cut, rstride=1, cstride=1, color='r',shade=True, alpha=0.3,
                       linewidth=0, antialiased=False)
#fig.colorbar(surf, shrink=0.5, aspect=5)
plt.title('Fig4: Threshold to dectect anomaly')
ax.set_xlabel('%名词', fontproperties=font)
ax.set_ylabel('%人名', fontproperties=font)
ax.view_init(20,45)
ax.set_xlim(0,1)
ax.set_ylim(0,1)
ax.set_zlim(0,5)

ax = fig.add_subplot(2,1,2,projection='3d')
mask = []
for item in ZZ:
    if item>prob_cut:
        mask.append(1)
    else:
        mask.append(0)
plt.title('Fig5: anomaly point(blue)')
ax.set_xlabel('%名词', fontproperties=font)
ax.set_ylabel('%人名', fontproperties=font)
ax.scatter3D(X_train[:,0],X_train[:,1],ZZ,c=mask)
ax.set_ylabel('%人名', fontproperties=font)
ax.view_init(20,45)
ax.set_xlim(0,1,emit=True)
ax.set_ylim(0,1,emit=True)
ax.set_zlim(0,5)

plt.show()
```

With the probability desity cut of 0.5, we detect 920 anomaly point out of 10,000 sample points. Then we check distribution of %tag_n and %tag_nr by excluding anomaly point, they show better gaussian distribution. 

```python
df_init_tag['labels'] = mask
size = df_init_tag.groupby(by='labels').count()

print '# of Amomaly contacts list custmer = %d, # of Normal contacts list custmer = %d' %(size.iloc[0,0],size.iloc[1,0])
ax_list = df_init_tag.hist('tag_n',by='labels',figsize=(20, 5),bins=np.linspace(0,1,50))
for ax_item in ax_list.flat:
    ax_item.set_xlim(0,1)
    text_title = lambda x: 'Anomaly' if x=='0' else 'Normal'
    ax_item.set_title( '%名词  '+text_title(ax_item.get_title()), fontproperties=font)

ax_list = df_init_tag.hist('tag_nr',by='labels',figsize=(20, 5),bins=np.linspace(0,1,50))
for ax_item in ax_list.flat:
    ax_item.set_xlim(0,1)
    text_title = lambda x: 'Anomaly' if x=='0' else 'Normal'
    ax_item.set_title( '%人名  '+text_title(ax_item.get_title()), fontproperties=font)

df_init_tag[df_init_tag['labels']==0].to_csv('F:\\data\\contacts\\temp_anomaly.csv',columns=['mobile_name'])
df_init_tag[df_init_tag['labels']==1].to_csv('F:\\data\\contacts\\temp_normal.csv',columns=['mobile_name'])
```
