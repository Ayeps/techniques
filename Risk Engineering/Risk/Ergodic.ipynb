When the gain multiple is worse than the loss multiple.  

You have 1000 in your kitty. You have an investment strategy by which you stake 

```python
You have 100 in a risky asset, this asset has a 50% chance to decrease to a value of 0.6 a 0.6
```

```python
### Because it is product it is ergodic
### as long as you don't go bankrupt
### you can keep on going.
import random
wealth = 100
u_mult = 2
d_mult = 0.5
time_series = []

for i in range(500):
    r = random.randint(0,1)
    if r==0:
        wealth = wealth*d_mult
    else:
        wealth = wealth*u_mult
    time_series.append(wealth)
```

```python
time_series
```

```python
### Because it is sum it is ergodic
### This is unfair as you don't get free
### bets you need collateral.

import random
wealth = 100
u_mult = 6
d_mult = 5
time_series = []

for i in range(500):
    r = random.randint(0,1)
    if r==0:
        wealth = wealth - d_mult
    else:
        wealth = wealth + u_mult
    time_series.append(wealth)
```

```python
time_series
```

```python
### Because it is sum and product it is non-ergodic
### If you keep reinvesting the full amount.

import random
wealth = 100
u_mult = 2
d_mult = 0.5
time_series = []

for i in range(500):
    r = random.randint(0,1)
    if r==0:
        wealth = wealth - wealth*d_mult
    else:
        wealth = wealth + wealth*u_mult
    time_series.append(wealth)
```

```python
time_series
```

```python
## The best is to take a small portion of your wealth to stake on bets.

### Because it is sum it is ergodic
import random
wealth = 100
time_series = []

for i in range(5000):
    r = random.randint(0,1)
    if r==0:
        wealth = wealth - (0.2*wealth)
    else:
        wealth = wealth + (0.2*wealth)
    time_series.append(wealth)
```
