```python
from os import listdir
import random
from os.path import isfile, join
mypath = "/Volumes/extra/NLP_Work/transc/"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

len(onlyfiles)

rrr = [ ra for ra in onlyfiles if ".DOC" not in ra ]

rrr = [ ra for ra in rrr if "._" not in ra ]
```

```python
zulu = [ rrr[i] for i in sorted(random.sample(range(len(rrr)), 500)) ]

data_text = []
for z in zulu:
    data_text.append(open(mypath + z).read())
```

```python
from gensim.models import word2vec
import nltk
# tokenize sentences in corpus
wpt = nltk.WordPunctTokenizer()
tokenized_corpus = [wpt.tokenize(document) for document in data_text]
## meant for multiple documents as you know
```

```python
# Set values for various parameters
feature_size = 100    # Word vector dimensionality  
window_context = 30          # Context window size                                                                                    
min_word_count = 1   # Minimum word count                        
sample = 1e-3   # Downsample setting for frequent words

w2v_model = word2vec.Word2Vec(tokenized_corpus, size=feature_size, 
                          window=window_context, min_count=min_word_count,
                          sample=sample, iter=50)
```

```python
w2v_model.save('w2v_model5.gensim')
```

```python
import _pickle as pickle
pickle.dump(tokenized_corpus, open('tokenized_corpus.pkl', 'wb'))
```

```python
# view similar words based on gensim's model
similar_words = {search_term: [item[0] for item in w2v_model.wv.most_similar([search_term], topn=100)]
                  for search_term in ['strategy', 'mergers', 'acquisition', 'pricing', 'execute', 'sales']}
similar_words
```

```python
# view similar words based on gensim's model
similar_words = {search_term: [item[0] for item in w2v_model.wv.most_similar([search_term], topn=5)]
                  for search_term in ['strategy', 'mergers', 'acquisition', 'pricing', 'execute', 'sales']}
similar_words
```

```python
## you can use these metrics PCA them and feed them into a ML model
## https://www.kdnuggets.com/2018/04/robust-word2vec-models-gensim.html

%matplotlib inline
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt 
import numpy as np

words = sum([[k] + v for k, v in similar_words.items()], [])
wvs = w2v_model.wv[words]

tsne = TSNE(n_components=2, random_state=0, n_iter=10000, perplexity=2)
np.set_printoptions(suppress=True)
T = tsne.fit_transform(wvs)
labels = words

plt.figure(figsize=(14, 8))
plt.scatter(T[:, 0], T[:, 1], c='orange', edgecolors='r')
for label, x, y in zip(labels, T[:, 0], T[:, 1]):
    plt.annotate(label, xy=(x+1, y+1), xytext=(0, 0), textcoords='offset points')
```
